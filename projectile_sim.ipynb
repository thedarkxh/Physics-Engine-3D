```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def plot_projectile_3d_with_metrics(u, theta, phi):
    g = 9.81
    theta_rad = np.radians(theta)
    phi_rad = np.radians(phi)

    t_flight = (2 * u * np.sin(theta_rad)) / g
    t = np.linspace(0, t_flight, num=500) # Higher resolution for distance accuracy

    # 3D Coordinates
    x = u * np.cos(theta_rad) * np.cos(phi_rad) * t
    y = u * np.cos(theta_rad) * np.sin(phi_rad) * t
    z = u * np.sin(theta_rad) * t - 0.5 * g * t**2

    # --- Calculation of Metrics ---
   # ... (previous code for x, y, z calculations)

    # 1. Final Displacement (Net)
    final_disp = np.sqrt(x[-1]**2 + y[-1]**2 + z[-1]**2)

    # 2. Total Distance (Arc Length)
    dx, dy, dz = np.diff(x), np.diff(y), np.diff(z)
    total_distance = np.sum(np.sqrt(dx**2 + dy**2 + dz**2))

    # --- ENHANCED PLOTTING WITH DYNAMIC AXES ---
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(x, y, z, color='#00FFCC', lw=3, label='Trajectory')
    ax.plot(x, y, 0, color='white', linestyle='--', alpha=0.3)

    # DYNAMIC SCALING: This ensures the graph grows with your theta and velocity
    # We find the max value among all axes to keep the plot "square" and proportional
    max_val = max(np.max(x), np.max(np.abs(y)), np.max(z), 10)

    ax.set_xlim(0, max_val)
    ax.set_ylim(-max_val/2, max_val/2)
    ax.set_zlim(0, max_val) # Now θ=85° won't hit the "ceiling"

    ax.set_title(f"3D Analysis: Velocity={u}m/s | Elevation={theta}\u00b0")
    # ... (rest of the plot formatting)

    plt.show()

    print(f"{'='*40}")
    print(f"DISPLACEMENT (Net): {final_disp:.2f} meters")
    print(f"TOTAL DISTANCE (Path): {total_distance:.2f} meters")
    print(f"{'='*40}")

interact(plot_projectile_3d_with_metrics,
         u=FloatSlider(value=30, min=10, max=60, step=1, description='Velocity:'),
         theta=FloatSlider(value=45, min=5, max=85, step=1, description='Elevation:'),
         phi=FloatSlider(value=0, min=-90, max=90, step=5, description='Azimuth:'))
```
